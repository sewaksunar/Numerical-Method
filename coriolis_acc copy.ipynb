```python
from sympy import *
from sympy.vector import *
from IPython.display import display, Math, Markdown
from sympy.physics.vector import *


init_printing(use_latex='mathjax')
print("SymPy", __import__('sympy').__version__, "loaded")
```

```python
from sympy import latex


# Scalar symbols 
s, t = symbols(r's t',real=True)   # position
s_dot, s_ddot = symbols(r'\dot{s}  \ddot{s}',real=True)   # speed & tangential accel
omega2 = symbols(r'\omega_2',real=True)   # angular speed of link 2
rho           = symbols(r'\rho',positive=True) # radius of curvature

# Vector symbols
# absolute frame of reference (inertial)
N = ReferenceFrame('N')

# link 2 frame of reference - rotates about N.x with angular velocity omega2
N2 = N.orientnew('N_2', 'Axis', [omega2*t, N.x])

# Angular velocity vector of frame N2 in frame N
omega2_vec = omega2 * N.x

# frame of reference at moving point P
cons_pho_s = symbols(r'C_{\rhos}')
s = Function('s')(t)
phi = s/rho + cons_pho_s

# link 2 frame of reference (tangential frame u)
# ux = tangential, uy = normal, uz = binormal
u = N2.orientnew('u', 'Axis', [phi, N2.x]) 

# Unit vectors
ut = u.x  # tangential
un = u.y  # normal
ub = u.z  # binormal

# Position vectors
x2, y2, z2 = symbols('x2 y2 z2')
R_C_2 = x2 * N2.x + y2 * N2.y + z2 * N2.z
display(Math(r"\vec R_{C_3} = " + latex(R_C_2)))

R_P_2 = R_C_2 - rho * u.y
display(Math(r"\vec R_{{P_3}/2} = " + latex(R_P_2)))

# Express in inertial frame N
R_P_N = R_P_2.express(N)
display(Math(r"\vec R_{{P_3}/N} = " + latex(R_P_N)))

# Velocity of P
V_P = s.diff(t, u) * u.x
display(Math(r"\vec V_{P} = " + latex(V_P)))

# Angular velocity of frame u relative to N2
vphi = phi * N2.x
dphi_dt = vphi.diff(t, N2)
display(Math(r"\dot{\phi} = " + latex(dphi_dt)))

# Time derivative of tangential unit vector in inertial frame
# dut/dt = (dut/dt)|_u + omega2_vec cross ut
# where (dut/dt)|_u is the time derivative in frame N2
dut_in_N2 = ut.diff(t, N2)
coriolis_term = omega2_vec.cross(ut)
dut_total = dut_in_N2 + coriolis_term

display(Math(r"\frac{d\vec u_t}{dt}\bigg|_{N_2} = " + latex(dut_in_N2)))
display(Math(r"\vec{\omega}_2 \times \vec{u}_t = " + latex(coriolis_term)))
display(Math(r"\frac{d\vec u_t}{dt}\bigg|_{N} = " + latex(dut_total)))
```
