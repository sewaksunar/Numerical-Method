```python
from sympy import *
x, y, z = symbols('x y z')
init_printing(use_unicode=True)
```

# Derivatives of function

```python
diff(cos(x), x)
```

# Multiple Derivatives of funtion

```python
diff(x**4, x, x, x) # Third derivative of x^4 with respect to x
```

```python
ex = exp(-x**2) 
ex.diff(x) # as a method
```

```python
expr = x**2 * sin(y)
expr.diff(x, y)
```

# Unevaluated derivative

```python
expr = exp(x*y*z)
deriv = Derivative(expr, x, y, y, z, 4)
deriv
```

# To evaluate unevaluated derivative

```python
deriv.doit()
```

# Integrals

```python
integrate(exp(-x), (x, 0, oo)) # define the limits integral from 0 to infinity
```

## double define intergal

```python
integrate(exp(-x**2 - y**2), (x, -oo, oo), (y, -oo, oo))
```

# Unevaluated integral

```python
expr = Integral(log(x)**2, x)
expr
```

```python
expr.doit()
```

```python
integ = Integral(sin(x**2), x)
integ
```

```python
integ.doit()
```

# Limits

```python
expr = Limit((cos(x) - 1)/x, x, 0)
expr
```
