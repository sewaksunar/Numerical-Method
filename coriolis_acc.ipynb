## General Formula: Acceleration in Rotating Reference Frames

For a point P moving in a rotating reference frame:

$$\vec{a}_{P/N} = \vec{a}_{P/B} + \vec{\alpha} \times \vec{r}_{P} + 2\vec{\omega}_B \times \vec{v}_{P/B} + \vec{\omega}_B \times (\vec{\omega}_B \times \vec{r}_{P})$$

Where:
- $\vec{a}_{P/N}$: acceleration in inertial frame N
- $\vec{a}_{P/B}$: acceleration relative to rotating frame B
- $\vec{\alpha}$: angular acceleration of frame B
- $\vec{\omega}_B$: angular velocity of frame B
- $\vec{r}_{P}$: position of P
- $\vec{v}_{P/B}$: velocity relative to frame B

**Key terms:**
1. **Relative acceleration**: $\vec{a}_{P/B}$ (in rotating frame)
2. **Angular acceleration effect**: $\vec{\alpha} \times \vec{r}_{P}$
3. **Coriolis acceleration**: $2\vec{\omega}_B \times \vec{v}_{P/B}$ ← **Critical for rotating frames**
4. **Centrifugal acceleration**: $\vec{\omega}_B \times (\vec{\omega}_B \times \vec{r}_{P})$ (radially outward)

**General Procedure:**
1. Define inertial frame N and rotating frame(s)
2. Express rotation: angle(t) and angular velocity $\vec{\omega}$
3. Compute position and velocity in rotating frame
4. Apply formula to get acceleration in inertial frame
5. Identify and separate each acceleration component
